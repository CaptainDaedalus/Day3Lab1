```python
import pandas as pd  # keywords are highlighted in green, other strings in red, etc.
import numpy as np
```

```python
#1. Show the DataFrame's shape.
mca = pd.read_csv('marketing_customer_analysis.csv')

display(mca.head())
display(mca.tail())
mca.shape
```

```python
#2. Standardize header names.
cols = []
for i in range(len(mca.columns)):
    cols.append(mca.columns[i].lower())
mca.columns = cols

mca.head()
```

```python
# 3.Which columns are numerical?
num_col = mca._get_numeric_data()
display(num_col)
```

```python
#4. Which columns are categorical?
cat_col = mca.select_dtypes(include=['object']).columns
display(cat_col)
```

```python
#5. Check and deal with NaN values.
show_nan = mca[mca.isna().any(axis=1)] #using isna() (or isnull()) to select all rows with NaN under an entire DataFrame
mca_value = mca.isnull().values.any() #check for NaN values is True
mca_sum = mca.isnull().sum().sum() #check for NaN sum total is 9254
display(show_nan)
display(mca_value)
display(mca_sum)

#I have a feeling this code is not super duper elegant
```
